## Работа с данными и построение карты

```python
#Выбор нужных колонок датасета - координаты, название и id станций

# Данные начальных станций
coord_start = df[["start_lat", "start_lng"]]
coord_start.columns = [["lat", "lng"]]

# Данные конечных станций
coord_end = df[["end_lat", "end_lng"]]
coord_end.columns = [["lat", "lng"]]
```

```python
# Склейка данных в 1 датасет
frames = [coord_start, coord_end]
coord = pd.concat(frames)
```

```python
# Удаление дублей
coord = coord.drop_duplicates()
```

```python
coord.info()
```

```python
coord_test = coord.head(100).values.tolist()
```

```python
from folium.plugins import MarkerCluster


chicago = folium.Map(location=[41.85, -87.65], zoom_start=11)

marker_cluster = MarkerCluster().add_to(chicago)



for i in coord_test:
    folium.CircleMarker(location=(i[0], i[1]), fill_color='#990066', color="gray", fill_opacity = 0.9).add_to(marker_cluster)

chicago
```
